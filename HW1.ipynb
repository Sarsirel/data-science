```python
# Задание №9
# Функция принимает строку и выводит ее в обратном порядке

def f9(tmp9=input('Введите текст: ')):
    a9 = ''
    for i in reversed(tmp9):
        a9 += i
    return a9

print(f9())
```

```python
# Задание №15
# Функция, принимающая список чисел и возвращающая максимальное

def f15(tmp_list=[0]):
    tmp_max = tmp_list[0]
    for i in tmp_list:
        if tmp_max < i:
            tmp_max = i
    return tmp_max

# контрольный пример
a15_list = [7, 14, 28, 1, 41, 53, 39]
a15_max = f15(a15_list)
a15_max
```
